# Question - 1

```python
# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset
data_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

# Step 3: The dataset is now stored in the variable 'users'
users = pd.read_csv(data_url, sep='|')

# Step 4: Calculate the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Calculate the male ratio per occupation and sort it
male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / x.count()).sort_values(ascending=False)

# Step 6: Calculate the minimum and maximum ages per occupation
age_range_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Step 7: Calculate the mean age per combination of occupation and sex
mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: Calculate the percentage of women and men per occupation
gender_percentage_per_occupation = users.groupby('occupation')['gender'].value_counts(normalize=True) * 100

# Print the results of each step
print("Mean Age per Occupation:\n", mean_age_per_occupation)
print("\nMale Ratio per Occupation (Most to Least):\n", male_ratio_per_occupation)
print("\nMinimum and Maximum Ages per Occupation:\n", age_range_per_occupation)
print("\nMean Age per Combination of Occupation and Sex:\n", mean_age_per_combination)
print("\nGender Percentage per Occupation:\n", gender_percentage_per_occupation)
```

```python
import pandas as pd

# Step 2: Import the dataset
data = pd.read_csv('wind.txt', delim_whitespace=True)

# Step 3: Convert the first 3 columns to datetime index
data['datetime'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=['Yr', 'Mo', 'Dy']).set_index('datetime')

# Step 4: Function to fix year 2061
def fix_year(year):
    if year > 2000:
        year -= 100
    return year

data.index = data.index.map(lambda x: x.replace(year=fix_year(x.year)))

# Step 5: Set the datetime index
data.index = pd.to_datetime(data.index, format='%y-%m-%d')

# Step 6: Count missing values for each location
missing_values_per_location = data.isnull().sum()
print("Missing values per location:")
print(missing_values_per_location)

# Step 7: Count non-missing values in total
non_missing_count = data.notnull().sum().sum()
print("\nTotal non-missing values:")
print(non_missing_count)

# Step 8: Calculate the mean windspeed over all locations and times
mean_windspeed = data.mean().mean()
print("\nMean windspeed over all locations and times:")
print(mean_windspeed)

# Step 9: Calculate min, max, mean, and std for each location
loc_stats = data.agg(['min', 'max', 'mean', 'std'])
print("\nLocation statistics:")
print(loc_stats)

# Step 10: Calculate min, max, mean, and std for each day
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])
print("\nDay statistics:")
print(day_stats)

# Step 11: Average windspeed in January for each location
january_avg = data[data.index.month == 1].groupby(data[data.index.month == 1].index.year).mean()
print("\nAverage windspeed in January for each year:")
print(january_avg)

# Step 12: Downsample to yearly frequency
yearly_data = data.resample('Y').mean()
print("\nYearly downsampled data:")
print(yearly_data)

# Step 13: Downsample to monthly frequency
monthly_data = data.resample('M').mean()
print("\nMonthly downsampled data:")
print(monthly_data)

# Step 14: Downsample to weekly frequency
weekly_data = data.resample('W-MON').mean()
print("\nWeekly downsampled data:")
print(weekly_data)

# Step 15: Calculate min, max, mean, and std for each week for the first 52 weeks
weekly_stats = data.resample('W-MON').agg(['min', 'max', 'mean', 'std']).iloc[:52]
print("\nWeekly statistics for the first 52 weeks:")
print(weekly_stats)
```

```python
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
print("First 10 entries:")
print(chipo.head(10))

# Step 5: Number of observations
num_observations = chipo.shape[0]
print("\nNumber of observations:", num_observations)

# Step 6: Number of columns
num_columns = chipo.shape[1]
print("\nNumber of columns:", num_columns)

# Step 7: Print the column names
print("\nColumn names:")
print(chipo.columns)

# Step 8: Index type
index_type = chipo.index
print("\nIndex type:", index_type)

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("\nMost-ordered item:", most_ordered_item)

# Step 10: Number of items ordered for most-ordered item
most_ordered_item_count = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("\nNumber of items ordered for the most-ordered item:", most_ordered_item_count)

# Step 11: Most ordered item in the choice_description column
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()
print("\nMost ordered item in the choice_description column:", most_ordered_choice)

# Step 12: Total items ordered
total_items_ordered = chipo['quantity'].sum()
print("\nTotal items ordered:", total_items_ordered)

# Step 13: Convert item price to float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Check item price type before and after conversion
print("\nItem price type before conversion:", chipo['item_price'].dtype)
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x))
print("Item price type after conversion:", chipo['item_price'].dtype)

# Step 14: Calculate revenue
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("\nTotal revenue:", revenue)

# Step 15: Number of orders made
num_orders = chipo['order_id'].nunique()
print("\nNumber of orders:", num_orders)

# Step 16: Average revenue per order
average_revenue_per_order = revenue / num_orders
print("\nAverage revenue amount per order:", average_revenue_per_order)

# Step 17: Number of different items sold
num_different_items = chipo['item_name'].nunique()
print("\nNumber of different items sold:", num_different_items)
```

# Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

# Calculate marriages and divorces per capita
data['Marriages_per_capita'] = data['Marriages'] / data['Population'] * 1000
data['Divorces_per_capita'] = data['Divorces'] / data['Population'] * 1000

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Marriages_per_capita'], label='Marriages per capita')
plt.plot(data['Year'], data['Divorces_per_capita'], label='Divorces per capita')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number per 1000')
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

# Calculate marriages and divorces per capita
data['Marriages_per_capita'] = data['Marriages'] / data['Population'] * 1000
data['Divorces_per_capita'] = data['Divorces'] / data['Population'] * 1000

# Select data for the years 1900, 1950, and 2000
selected_years = [1900, 1950, 2000]
selected_data = data[data['Year'].isin(selected_years)]

# Create a vertical bar chart
plt.figure(figsize=(8, 6))
bar_width = 0.15
bar_positions = range(len(selected_years))
plt.bar(bar_positions, selected_data['Marriages_per_capita'], width=bar_width, label='Marriages per capita')
plt.bar([pos + bar_width for pos in bar_positions], selected_data['Divorces_per_capita'], width=bar_width, label='Divorces per capita')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number per 1000')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xticks([pos + bar_width / 2 for pos in bar_positions], selected_years)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

# Question - 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("actor_kill_counts.csv")

# Sort the actors by kill count
sorted_data = data.sort_values(by='Count', ascending=True)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(sorted_data['Actor'], sorted_data['Count'])

# Add labels and title
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')
plt.yticks(rotation=45)  # Rotate actor names for better visibility

# Annotate each bar with the actor's name
for i, count in enumerate(sorted_data['Count']):
    plt.text(count + 5, i, str(count), va='center')  # Adding count value next to each bar

# Show the plot
plt.tight_layout()
plt.show()
```

# Question 9
Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("roman-emperor-reigns.csv")

# Calculate the count of emperors in each cause of death category
cause_counts = data['Cause_of_Death'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(cause_counts, labels=cause_counts.index, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Fraction of Roman Emperors by Cause of Death')

# Equal aspect ratio ensures that the pie chart is circular
plt.axis('equal')

# Show the plot
plt.show()
```

# Question 10
Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("arcade-revenue-vs-cs-doctorates.csv")

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Total Arcade Revenue (billions)'], data['Computer Science Doctorates Awarded (US)'],
            c=data['Year'], cmap='viridis', s=100, edgecolors='k', alpha=0.7)

# Add labels and title
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.title('Relationship between Arcade Revenue and CS Doctorates Awarded (2000-2009)')
plt.colorbar(label='Year')

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
